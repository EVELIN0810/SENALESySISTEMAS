##   **Representaciones de Fourier: Definiciones y comparaciones**

Las herramientas matem√°ticas basadas en la Transformada de Fourier permiten analizar se√±ales en el dominio de la frecuencia. A continuaci√≥n se describen sus principales variantes:

---

###  **Serie de Fourier (exponencial, trigonom√©trica y compacta)**

La **serie de Fourier** permite representar una **se√±al peri√≥dica** como una suma infinita de senoidales. Existen tres formas comunes de representaci√≥n:

- **Trigonom√©trica**: combina senos y cosenos de diferentes frecuencias.

  $$
  x(t) = a_0 + \sum_{n=1}^{\infty} [a_n \cos(n\omega_0 t) + b_n \sin(n\omega_0 t)]
  $$

- **Exponencial compleja**:

  $$
  x(t) = \sum_{n=-\infty}^{\infty} C_n e^{j n \omega_0 t}
  $$

- **Forma compacta (fasorial)**: considera magnitud y fase como coeficientes complejos.

Esta herramienta **solo aplica a se√±ales peri√≥dicas**. Su espectro es **discreto**, compuesto por l√≠neas en las frecuencias arm√≥nicas de la se√±al.

---

### **Transformada de Fourier (FT)**

La **Transformada de Fourier** extiende la idea de la serie de Fourier a se√±ales **no peri√≥dicas** (en tiempo continuo):

$$
X(f) = \int_{-\infty}^{\infty} x(t) e^{-j 2 \pi f t} dt
$$

$$
x(t) = \int_{-\infty}^{\infty} X(f) e^{j 2 \pi f t} df
$$

- Aplica a se√±ales **continuas** y **no peri√≥dicas**.
- El espectro resultante es **continuo**, abarcando todas las frecuencias.

---

### **Transformada de Fourier en Tiempo Discreto (DTFT)**

La **DTFT** se usa para representar se√±ales **discretas en tiempo** pero de duraci√≥n **infinita**:

$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j \omega n}
$$

- La DTFT produce un **espectro continuo** pero **peri√≥dico** en $$\omega$$, con periodo $$2\pi$$
- No es computable directamente, pero es fundamental para el an√°lisis te√≥rico de se√±ales digitales.

---

### **Transformada Discreta de Fourier (DFT)**

La **DFT** es una versi√≥n computable de la DTFT que trabaja sobre se√±ales **discretas y de duraci√≥n finita** (N muestras):

$$
X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2\pi}{N}kn}, \quad k = 0, 1, ..., N-1
$$

- Su espectro es **discreto** y contiene **N componentes de frecuencia**.
- Se aplica en contextos digitales y permite el an√°lisis num√©rico en computadores.

---

### **Comparaci√≥n general**

| Herramienta          | Tipo de se√±al            | Espectro           | Periodicidad en el tiempo | Es computable | Aplicaci√≥n principal                      |
|----------------------|--------------------------|--------------------|----------------------------|---------------|-------------------------------------------|
| Serie de Fourier     | Continua, peri√≥dica      | Discreto           | Peri√≥dica                 | No            | An√°lisis de se√±ales peri√≥dicas            |
| Transformada de Fourier (FT) | Continua, no peri√≥dica | Continuo           | No peri√≥dica              | No            | An√°lisis espectral general en tiempo continuo |
| DTFT                 | Discreta, infinita       | Continuo y peri√≥dico en frecuencia | No peri√≥dica | No            | An√°lisis te√≥rico de se√±ales digitales     |
| DFT                  | Discreta, finita         | Discreto           | Finita                     | S√≠            | Procesamiento digital de se√±ales (DSP)    |

---

##  **Algoritmo Fast Fourier Transform (FFT)**

La **FFT (Fast Fourier Transform)** es un algoritmo eficiente para calcular la DFT, ampliamente usado en la pr√°ctica.

### Utilidad

- Reduce el tiempo de procesamiento.
- Es el n√∫cleo de herramientas como an√°lisis de espectros, filtros digitales, compresi√≥n de audio (MP3), procesamiento de imagen (JPEG), etc.

---

### Definici√≥n de DFT:

$$
X[k] = \sum_{n=0}^{N-1} x[n] \cdot e^{-j\frac{2\pi}{N}kn}
$$

- Esta f√≥rmula requiere \(N^2\) operaciones para calcular todos los \(X[k]\), lo cual no es eficiente para grandes valores de \(N\).

---

### FFT: Idea general

El algoritmo FFT **divide y conquista** el c√°lculo de la DFT:

- Se separan las muestras de la se√±al en √≠ndices **pares** e **impares**.
- Se calcula la DFT de cada parte de forma recursiva.
- Se recombinan los resultados utilizando simetr√≠as del exponencial complejo.

### Complejidad computacional

| M√©todo | N√∫mero de operaciones | Notaci√≥n Big-O |
|--------|------------------------|----------------|
| DFT directa | $$ N^2 $$               | $$ O(N^2) $$     |
| FFT (Cooley-Tukey) | $$ N \log_2 N $$         | $$ O(N \log N) $$ |

---

### Conclusi√≥n

- La **serie de Fourier** se utiliza para se√±ales peri√≥dicas.
- La **FT** y **DTFT** son an√°lisis de frecuencia para se√±ales continuas y discretas no peri√≥dicas, respectivamente.
- La **DFT** es la versi√≥n computacional que permite analizar se√±ales en computadores.
- La **FFT** es el algoritmo que hace que el uso de la DFT sea eficiente y viable en tiempo real o sistemas embebidos.



# Punto 2 ‚Äî Transformada de Fourier desde la definici√≥n

Encuentre la **funci√≥n de densidad espectral** (transformada de Fourier) de las siguientes se√±ales **sin aplicar propiedades**:

1. $$ x(t) = e^{-a|t|}, \quad a \in \mathbb{R}^{+} $$  
2. $$ x(t) = \cos(\omega_c t), \quad \omega_c \in \mathbb{R} $$
3. $$ x(t) = \sin(\omega_s t), \quad \omega_s \in \mathbb{R} $$
4. $$ x(t) = f(t) \cos(\omega_c t), \quad f(t) \in \mathbb{R} \text{ o } \mathbb{C}, \ \omega_c \in \mathbb{R} $$
5. $$ x(t) = e^{-a t^2}, \quad a \in \mathbb{R}^{+} $$  
6. $$ x(t) = A \cdot \text{rect}_d(t), \quad A, d \in \mathbb{R} $$

---

## 1. Se√±al: $$ x(t) = e^{-a|t|} $$

**Paso 1:** Separar la funci√≥n por tramos:

$$
x(t) =
\begin{cases}
e^{-a t}, & t \geq 0 \\
e^{a t},  & t < 0
\end{cases}
$$

**Paso 2:** Aplicar la definici√≥n:

$$
X(f) = \int_{-\infty}^{\infty} e^{-a|t|} e^{-j 2\pi f t} dt
= \int_{-\infty}^{0} e^{a t} e^{-j 2\pi f t} dt + \int_{0}^{\infty} e^{-a t} e^{-j 2\pi f t} dt
$$

**Primer integral:**

$$
\int_{-\infty}^{0} e^{(a - j2\pi f)t} dt
= \left[ \frac{e^{(a - j2\pi f)t}}{a - j2\pi f} \right]_{-\infty}^{0}
= \frac{1}{a - j2\pi f}
$$

**Segunda integral:**

$$
\int_{0}^{\infty} e^{-(a + j2\pi f)t} dt
= \left[ \frac{e^{-(a + j2\pi f)t}}{-(a + j2\pi f)} \right]_0^{\infty}
= \frac{1}{a + j2\pi f}
$$

**Resultado final:**

$$
X(f) = \frac{1}{a - j2\pi f} + \frac{1}{a + j2\pi f}
= \frac{2a}{a^2 + (2\pi f)^2}
$$

---

## 2. Se√±al: $$ x(t) = \cos(\omega_c t) $$

**Paso 1:** Escribir el coseno en forma exponencial:

$$
\cos(\omega_c t) = \frac{1}{2} \left( e^{j \omega_c t} + e^{-j \omega_c t} \right)
$$

**Paso 2:** Sustituir en la definici√≥n:

$$
X(f) = \int_{-\infty}^{\infty} \cos(\omega_c t) e^{-j2\pi f t} dt
= \frac{1}{2} \int_{-\infty}^{\infty} e^{j \omega_c t} e^{-j2\pi f t} dt + \frac{1}{2} \int_{-\infty}^{\infty} e^{-j \omega_c t} e^{-j2\pi f t} dt
$$

$$
= \frac{1}{2} \int_{-\infty}^{\infty} e^{-j2\pi (f - \omega_c/2\pi) t} dt + \frac{1}{2} \int_{-\infty}^{\infty} e^{-j2\pi (f + \omega_c/2\pi) t} dt
$$

Estas integrales son de la forma $$\int e^{-j2\pi a t} dt $$ que da $$ \delta(a) $$

$$
X(f) = \frac{1}{2} \delta\left(f - \frac{\omega_c}{2\pi}\right) + \frac{1}{2} \delta\left(f + \frac{\omega_c}{2\pi}\right)
$$

---

## 3. Se√±al: $$ x(t) = \sin(\omega_s t) $$

**Paso 1:** Forma exponencial:

$$
\sin(\omega_s t) = \frac{1}{2j} \left( e^{j \omega_s t} - e^{-j \omega_s t} \right)
$$

**Paso 2:** Sustituir:

$$
X(f) = \frac{1}{2j} \left[ \int_{-\infty}^{\infty} e^{j \omega_s t} e^{-j2\pi f t} dt - \int_{-\infty}^{\infty} e^{-j \omega_s t} e^{-j2\pi f t} dt \right]
$$

$$
= \frac{1}{2j} \left[ \delta\left(f - \frac{\omega_s}{2\pi}\right) - \delta\left(f + \frac{\omega_s}{2\pi}\right) \right]
$$

---

## 4. Se√±al: $$ x(t) = f(t) \cos(\omega_c t) $$

**Paso 1:** Usar identidad del coseno:

$$
x(t) = \frac{1}{2} f(t) e^{j \omega_c t} + \frac{1}{2} f(t) e^{-j \omega_c t}
$$

**Paso 2:** Aplicar definici√≥n:

$$
X(f) = \int_{-\infty}^{\infty} x(t) e^{-j2\pi f t} dt
= \frac{1}{2} \int_{-\infty}^{\infty} f(t) e^{-j2\pi(f - \omega_c/2\pi) t} dt + \frac{1}{2} \int_{-\infty}^{\infty} f(t) e^{-j2\pi(f + \omega_c/2\pi) t} dt
$$

Reconocemos que son las transformadas de: $$ f(t) $$ evaluadas en frecuencias desplazadas:

$$
X(f) = \frac{1}{2} F\left(f - \frac{\omega_c}{2\pi}\right) + \frac{1}{2} F\left(f + \frac{\omega_c}{2\pi}\right)
$$

---

## 5. Se√±al: $$ x(t) = e^{-a t^2} $$

**Paso 1:** Aplicar definici√≥n:

$$
X(f) = \int_{-\infty}^{\infty} e^{-a t^2} e^{-j2\pi f t} dt
$$

**Paso 2:** Resolver esta integral gaussiana con t√©rmino lineal:

Esta es una integral conocida:

$$
X(f) = \sqrt{\frac{\pi}{a}} e^{-\frac{(\pi f)^2}{a}}
$$

---

## 6. Se√±al: $$ x(t) = A \cdot \text{rect}_d(t) $$

**Paso 1:** Definici√≥n de la funci√≥n rectangular:

$$
\text{rect}_d(t) =
\begin{cases}
1, & |t| \leq \frac{d}{2} \\
0, & \text{en otro caso}
\end{cases}
$$

**Paso 2:** Evaluar la integral:

$$
X(f) = \int_{-\infty}^{\infty} A \cdot \text{rect}_d(t) \cdot e^{-j2\pi f t} dt
= A \int_{-d/2}^{d/2} e^{-j2\pi f t} dt
$$

$$
= A \left[ \frac{e^{-j2\pi f t}}{-j2\pi f} \right]_{-d/2}^{d/2}
= A \cdot \frac{\sin(\pi f d)}{\pi f}
= A d \cdot \text{sinc}(f d)
$$

Donde:

$$
\text{sinc}(x) = \frac{\sin(\pi x)}{\pi x}
$$

## a) $$ \mathcal{F} \{ e^{-j\omega_1 t} \cos(\omega_c t) \} $$

### Paso 2: Usar tabla:

Sabemos que:

- $$ \mathcal{F}\{ u(t) \} = \frac{1}{j2\pi f} + \frac{1}{2} \delta(f) $$

#  Punto 4 ‚Äî Modulaci√≥n AM por detecci√≥n coherente (explicaci√≥n te√≥rica)

## ¬øQu√© es la modulaci√≥n AM?

En **modulaci√≥n de amplitud (AM)**, una se√±al senoidal (portadora) cambia su **amplitud** en funci√≥n de una **se√±al mensaje** \( m(t) \). La se√±al modulada es:

$$
y(t) = \left(1 + \mu \cdot m(t)\right) \cdot \cos(2\pi f_c t)
$$

Donde:

- $$ m(t) $$: se√±al mensaje (normalizada entre -1 y 1)
- $$ f_c $$: frecuencia de la portadora (Hz)
- $$ \mu $$: √≠ndice de modulaci√≥n (0 < Œº ‚â§ 1)
- $$ y(t) $$: se√±al modulada

---

## Detecci√≥n coherente (demodulaci√≥n AM)

Para recuperar la se√±al mensaje, se multiplica la se√±al AM por una **portadora local sincronizada** (misma frecuencia y fase), y luego se aplica un filtro pasa bajas (idealmente):

1. Multiplicaci√≥n:

$$
y(t) \cdot \cos(2\pi f_c t) = \left(1 + \mu m(t)\right) \cos^2(2\pi f_c t)
$$

2. Usando $$ \cos^2(x) = \frac{1}{2}(1 + \cos(2x)) $$, se obtiene:

$$
= \frac{1}{2} \left(1 + \mu m(t)\right) + \frac{1}{2} \left(1 + \mu m(t)\right) \cos(4\pi f_c t)
$$

3. Luego de filtrar la componente de alta frecuencia $$ \cos(4\pi f_c t) $$:

$$
\text{Se√±al recuperada:} \quad \frac{1}{2} (1 + \mu m(t))
$$

Finalmente, se puede restar la componente continua y escalar para recuperar $$ m(t) $$.

---

## Condiciones para detecci√≥n coherente exitosa

- $$ \mu \cdot |m(t)| < 1 $$: evita sobre-modulaci√≥n
- La portadora del receptor debe estar **en fase y frecuencia sincronizada**
- Se requiere un buen **filtro pasa bajas**

---

## An√°lisis espectral

### Se√±al portadora:
$$
c(t) = \cos(2\pi f_c t) \Rightarrow C(f) = \frac{1}{2} \left[ \delta(f - f_c) + \delta(f + f_c) \right]
$$

---

### Se√±al mensaje:

1. **Pulso rectangular** de duraci√≥n $$ T $$:

$$
m(t) = \text{rect}\left(\frac{t}{T}\right) \Rightarrow M(f) = T \cdot \text{sinc}(fT)
$$

2. **Coseno de baja frecuencia** $$ f_m $$:

$$
m(t) = \cos(2\pi f_m t) \Rightarrow M(f) = \frac{1}{2} \left[ \delta(f - f_m) + \delta(f + f_m) \right]
$$

---

### Se√±al modulada AM:

El espectro de:

$$
y(t) = (1 + \mu m(t)) \cos(2\pi f_c t)
$$

Contiene:

- Una componente central en $$ f = f_c $$ (la portadora)
- Dos bandas laterales en $$ f_c \pm f_m $$ (para se√±al sinusoidal)
- En general, si $$ M(f) $$ es el espectro de $$ m(t) $$, entonces:

$$
Y(f) = \frac{1}{2} \left[ \delta(f - f_c) + \delta(f + f_c) \right]
+ \frac{\mu}{2} \left[ M(f - f_c) + M(f + f_c) \right]
$$

Esto significa que la modulaci√≥n AM desplaza el espectro del mensaje hacia ambos lados de la portadora.

---

## üìä Resumen

| Componente       | Tiempo                                           | Frecuencia                                  |
|------------------|--------------------------------------------------|---------------------------------------------|
| Mensaje $$ m(t) $$ | Rectangular o Cosenoide                         | sinc o dos deltas (seg√∫n el caso)           |
| Portadora         | Coseno de alta frecuencia                        | dos deltas en $$ \pm f_c $$                 |
| AM $$ y(t) $$     | Producto del mensaje con la portadora            | Portadora + bandas laterales desplazadas    |
| Recuperaci√≥n      | Multiplicaci√≥n por portadora + filtro pasa bajas | Se obtiene $$ m(t) $$ a partir de $$ y(t) $$|

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
from ipywidgets import interact, FloatSlider, Dropdown

# üîç Funci√≥n para graficar tiempo y frecuencia
def plot_time_and_freq(signal, title, fs, t, color='b'):
    N = len(signal)
    freq = rfftfreq(N, 1/fs)
    spectrum = np.abs(rfft(signal)) * 2 / N

    fig, axs = plt.subplots(1, 2, figsize=(12, 4))
    axs[0].plot(t * 1000, signal, color)
    axs[0].set_title(f"{title} en el tiempo")
    axs[0].set_xlabel("Tiempo [ms]")
    axs[0].grid()

    axs[1].plot(freq / 1000, spectrum, color)
    axs[1].set_title(f"{title} en frecuencia")
    axs[1].set_xlabel("Frecuencia [kHz]")
    axs[1].set_xlim(0, 10)
    axs[1].grid()

    plt.tight_layout()
    plt.show()

# üéõ Funci√≥n principal interactiva
def simular_am(tipo_mensaje, mu):
    # ‚öôÔ∏è Par√°metros base
    fs = 50000
    T = 0.02
    t = np.linspace(0, T, int(fs * T), endpoint=False)
    fc = 3000  # Hz

    # üìù Definici√≥n del mensaje
    if tipo_mensaje == "rectangular":
        ancho = 0.01
        m = np.where(np.abs(t - T/2) <= ancho / 2, 1.0, 0.0)
    elif tipo_mensaje == "coseno":
        fm = 100
        m = np.cos(2 * np.pi * fm * t)
    elif tipo_mensaje == "seno":
        fm = 100
        m = np.sin(2 * np.pi * fm * t)
    else:
        raise ValueError("Tipo de mensaje no v√°lido.")

    # Normalizar
    m = m / np.max(np.abs(m))

    # üîä Portadora
    c = np.cos(2 * np.pi * fc * t)

    # üîµ AM
    y = (1 + mu * m) * c

    # üü¢ Detecci√≥n coherente
    demod = y * c

    # Filtro pasa bajas en frecuencia
    def filtro_pasabajo_fft(signal, fs, fcut):
        N = len(signal)
        S = rfft(signal)
        freqs = rfftfreq(N, 1/fs)
        S_filtered = np.where(freqs <= fcut, S, 0)
        return np.fft.irfft(S_filtered)

    m_rec = filtro_pasabajo_fft(demod, fs, 1500)

    # üìä Gr√°ficas
    plot_time_and_freq(m, "Se√±al mensaje", fs, t, 'gray')
    plot_time_and_freq(c, "Portadora", fs, t, 'orange')
    plot_time_and_freq(y, "AM modulada", fs, t, 'blue')
    plot_time_and_freq(demod, "Detecci√≥n coherente", fs, t, 'green')
    plot_time_and_freq(m_rec, "Mensaje recuperado", fs, t, 'red')

# ‚öôÔ∏è Widgets interactivos
interact(
    simular_am,
    tipo_mensaje=Dropdown(
        options=["rectangular", "coseno", "seno"],
        value="coseno",
        description="Mensaje:"
    ),
    mu=FloatSlider(value=0.8, min=0.1, max=1.0, step=0.1, description="Œº")
);
```

# Punto 5 ‚Äî Aplicaci√≥n en comunicaciones: Modulaci√≥n AM con audio real

---



1. Sea la se√±al portadora:  
   $$
   c(t) = A_c \cos(2\pi f_c t)
   $$

2. Sea la se√±al mensaje:  
   $$
   m(t) \in \mathbb{R}
   $$

3. Se√±al modulada:  
   $$
   y(t) = \left(1 + \frac{m(t)}{A_c} \right) \cdot c(t)
   $$

4. Se solicita:

- Determinar el **espectro en frecuencia** de $$ y(t) $$
- Simular una se√±al AM con un **fragmento de audio** (canci√≥n)
- Utilizar un **√≠ndice de modulaci√≥n $$ \mu = 1 $$**
- Graficar en **tiempo y frecuencia**:
  - Mensaje
  - Portadora
  - Se√±al modulada
- Reproducir el audio de cada etapa
- Asumir un **demodulador coherente**, determinar el **espectro en cada etapa**
- Utilizar **filtrado ideal espectral** con `FFT` en la demodulaci√≥n

---

##  Se√±al AM con mensaje real

### Se√±al modulada:

La expresi√≥n:

$$
y(t) = \left(1 + \frac{m(t)}{A_c} \right) \cdot A_c \cos(2\pi f_c t) = A_c \cos(2\pi f_c t) + m(t) \cos(2\pi f_c t)
$$

muestra que la modulaci√≥n AM es una suma de:

1. La portadora pura: $$ A_c \cos(2\pi f_c t) $$
2. La se√±al mensaje multiplicada por la portadora: $$ m(t) \cos(2\pi f_c t) $$

---

### Espectro de $$ y(t) $$

El espectro de $$ y(t) $$ incluye:

- Un **delta** en $$ f = \pm f_c $$ (la portadora)
- Dos **r√©plicas del espectro de $$ m(t) $$** centradas en $$ \pm f_c $$, generadas por $$ m(t) \cdot \cos(2\pi f_c t) $$

---

### Ejemplo pr√°ctico:

Para este punto, se debe:

- Cargar un fragmento de canci√≥n (por ejemplo, archivo `.wav`)
- Normalizar y centrar $$ m(t) $$
- Elegir $$ A_c = \max(|m(t)|) $$ para garantizar $$ \mu = 1 $$
- Modulaci√≥n AM como:

  $$
  y(t) = \left(1 + \frac{m(t)}{A_c} \right) \cdot \cos(2\pi f_c t)
  $$

---

### Demodulaci√≥n coherente:

1. Multiplicaci√≥n con la misma portadora:

$$
y(t) \cdot \cos(2\pi f_c t)
= \left(1 + \frac{m(t)}{A_c} \right) \cos^2(2\pi f_c t)
$$

2. Aplicando identidad trigonom√©trica:

$$
= \frac{1}{2} \left(1 + \frac{m(t)}{A_c} \right)
+ \frac{1}{2} \left(1 + \frac{m(t)}{A_c} \right) \cos(4\pi f_c t)
$$

3. Aplicando filtro pasa bajas:

- Se elimina la componente de alta frecuencia
- Se obtiene:

$$
\text{Se√±al recuperada: } \frac{1}{2} \left(1 + \frac{m(t)}{A_c} \right)
$$

- Se puede escalar y restar la continua para obtener \( m(t) \)

---

##  Etapas esperadas

| Etapa                             | Se√±al                  | Transformada en frecuencia                |
|----------------------------------|------------------------|-------------------------------------------|
| Mensaje original                 | $$ m(t) $$            | $$ M(f) $$ (amplitud banda base)          |
| Portadora                        | $$ \cos(2\pi f_c t) $$ | $$ \delta(f \pm f_c) $$                        |
| Se√±al modulada AM                | $$ y(t) $$            | Portadora + bandas laterales $$ M(f \pm f_c) $$ |
| Multiplicaci√≥n con portadora     | $$ y(t) \cdot c(t) $$ | Componentes en DC y $$ 2f_c $$            |
| Se√±al demodulada (filtrada)      | $$ m(t) $$            | Recuperaci√≥n de $$ M(f) $$                |

---

### Flujo de trabajo

1. Descargar 5 s de audio desde YouTube.
2. Extraer la porci√≥n del segundo 20 al 25.
3. Aplicar modulaci√≥n AM con √≠ndice de modulaci√≥n \( \mu = 1 \).
4. Graficar en tiempo y frecuencia:
   - Mensaje $$ m(t) $$
   - Portadora $$ c(t) $$
   - Se√±al modulada $$ y(t) $$
5. Reproducir los tres audios para validar el efecto.

---

A continuaci√≥n se implementar√° un c√≥digo en Python para cumplir todos estos pasos.


```python
# ‚úÖ Paso 1: Instalar yt-dlp y ffmpeg si no est√°n
!pip install -q yt-dlp
!apt-get -qq install ffmpeg

# ‚úÖ Paso 2: Descargar audio completo de YouTube
import os

# Cambia esta variable con tu enlace:
youtube_url = "https://www.youtube.com/watch?v=_6HpI5i84w8"
output_file = "audio_fragment.wav"

# Eliminar archivo anterior si existe
if os.path.exists(output_file):
    os.remove(output_file)

# Descargar el audio completo y convertirlo a WAV
!yt-dlp -x --audio-format wav -o "{output_file}" "{youtube_url}"

# ‚úÖ Paso 3: Procesamiento en Python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fft import rfft, rfftfreq, irfft
from IPython.display import Audio, display

# ‚úÖ Cargar el audio
fs, m = wavfile.read("audio_fragment.wav")

# Normalizar y convertir a mono si es necesario
m = m.astype(np.float32)
m /= np.max(np.abs(m))
if m.ndim > 1:
    m = m[:, 0]

# ‚úÖ Verificar duraci√≥n total
duracion_total = len(m) / fs
print(f"Duraci√≥n total del audio: {duracion_total:.2f} segundos")

# ‚úÖ Extraer segmento del segundo 20 al 25
inicio_seg = 20
fin_seg = 25
if duracion_total < fin_seg:
    raise ValueError(f"‚ùå El audio solo dura {duracion_total:.2f} s. No se puede extraer del segundo {inicio_seg} al {fin_seg}.")

m = m[inicio_seg * fs : fin_seg * fs]
duracion = fin_seg - inicio_seg
N = len(m)
t = np.linspace(0, duracion, N, endpoint=False)

# ‚úÖ Modulaci√≥n AM
fc = 10000  # Hz
Ac = 1.0
mu = 1.0
c = Ac * np.cos(2 * np.pi * fc * t)
y = (1 + mu * m) * c

# ‚úÖ Demodulaci√≥n coherente
y_demod = y * np.cos(2 * np.pi * fc * t)

# ‚úÖ Filtro pasa bajas ideal
def filtro_pasa_bajas(signal, fs, fcorte):
    N = len(signal)
    F = rfft(signal)
    f = rfftfreq(N, d=1/fs)
    filtro = np.where(f <= fcorte, 1, 0)
    F_filtrado = F * filtro
    return irfft(F_filtrado, n=N)

m_recuperada = filtro_pasa_bajas(y_demod, fs, fcorte=4000)

# ‚úÖ Escuchar las se√±ales
print("üîä Se√±al mensaje original (m(t))")
display(Audio(m, rate=fs))

print("üîä Se√±al modulada (y(t))")
display(Audio(y, rate=fs))

print("üîä Despu√©s de multiplicaci√≥n coherente")
display(Audio(y_demod, rate=fs))

print("üîä Se√±al recuperada (mensaje demodulado)")
display(Audio(m_recuperada, rate=fs))
```

```python
from scipy.fft import rfft, rfftfreq

# ‚úÖ Funci√≥n para graficar espectro de una se√±al
def graficar_espectro(senal, fs, titulo):
    f = rfftfreq(len(senal), d=1/fs)            # Eje de frecuencias
    S = np.abs(rfft(senal)) / len(senal)        # Magnitud espectral normalizada
    plt.plot(f, S)
    plt.title(titulo)
    plt.xlabel("Frecuencia (Hz)")
    plt.ylabel("Magnitud")
    plt.xlim(0, fs // 2)                         # Hasta Nyquist

# ‚úÖ Graficar los espectros
plt.figure(figsize=(14, 6))

plt.subplot(3, 1, 1)
graficar_espectro(y, fs, "Espectro de $y(t)$ (modulada)")

plt.subplot(3, 1, 2)
graficar_espectro(y_demod, fs, "Espectro despu√©s de multiplicaci√≥n coherente")

plt.subplot(3, 1, 3)
graficar_espectro(m_recuperada, fs, "Espectro del mensaje recuperado")

plt.tight_layout()
plt.show()
```

# Punto 6 ‚Äî Aplicaci√≥n en circuitos el√©ctricos: THD y Factor de Potencia

---


- Explicar qu√© es la **Distorsi√≥n Total Arm√≥nica (THD)** y c√≥mo se calcula mediante la **FFT**
- Explicar el **Factor de Potencia (PF)** y su relaci√≥n con la distorsi√≥n
- Simular y comparar el comportamiento de un **rectificador de onda completa** con:
  1. Carga resistiva pura $$ R $$
  2. Carga en serie $$ R + C $$

---

## ¬øQu√© es el THD?

La **Distorsi√≥n Total de Arm√≥nicos (THD)** es una medida de cu√°nta energ√≠a est√° contenida en las componentes arm√≥nicas de una se√±al no senoidal respecto a su componente fundamental.

###  Forma general:

$$
\text{THD} = \frac{\sqrt{V_2^2 + V_3^2 + \cdots + V_n^2}}{V_1}
$$

Donde:

- $$ V_1 $$ componente RMS de la **frecuencia fundamental**
- $$ V_k $$ componente RMS de la **k-√©sima arm√≥nica** para $$ k \geq 2 $$

###  Paso a paso:

1. Obtener la transformada r√°pida de Fourier (FFT) de la se√±al:

   $$
   X[k] = \text{FFT}[x[n]]
   $$

2. Calcular la magnitud de cada componente:

   $$
   |X[k]| = \text{magnitud de la componente } k
   $$

3. Determinar el √≠ndice $$ k_1 $$ que corresponde a la frecuencia fundamental $$ f_1 $$. Luego:

   - $$ V_1 = |X[k_1]| $$
   - $$ V_2 = |X[2k_1]| $$
   - $$ V_3 = |X[3k_1]| $$, etc.

4. Aplicar la f√≥rmula:

   $$
   \text{THD} = \frac{\sqrt{\sum_{k=2}^{N} |X[k \cdot k_1]|^2}}{|X[k_1]|}
   $$

---

## ¬øQu√© es el Factor de Potencia (PF)?

El **Factor de Potencia Total (TPF)** tiene en cuenta tanto el desfase (como en cargas lineales) como la distorsi√≥n arm√≥nica.

Se define como:

$$
\text{TPF} = \cos(\phi_1) \cdot \frac{1}{\sqrt{1 + \text{THD}^2}}
$$

Donde:

- $$ \phi_1 $$ √°ngulo de desfase entre **tensi√≥n** e **intensidad** en la **componente fundamental**
- $$ \cos(\phi_1) $$ llamado **DPF** (Displacement Power Factor)
- $$ \frac{1}{\sqrt{1 + \text{THD}^2}} $$ llamado **DF** (Distortion Factor)

---

### Desglosado:

#### 1. DPF:

$$
\text{DPF} = \cos(\phi_1)
$$

Donde $$ \phi_1 $$ se obtiene como la diferencia de fase entre la tensi√≥n y corriente **en la frecuencia fundamental**, calculada desde la FFT:

$$
\phi_1 = \angle I_1 - \angle V_1
$$

#### 2. DF:

$$
\text{DF} = \frac{1}{\sqrt{1 + \text{THD}^2}}
$$

#### 3. TPF:

Finalmente,

$$
\text{TPF} = \text{DPF} \cdot \text{DF}
$$

---

##  Ejemplo: Rectificador de onda completa

###  Caso 1: Carga Resistiva $$ R $$

- La corriente es pulsada pero no se desfasa
- Tiene muchas arm√≥nicas ‚Üí **THD alto**
- Como no hay desfase:

$$
\text{DPF} = 1
\Rightarrow \text{TPF} = \frac{1}{\sqrt{1 + \text{THD}^2}} < 1
$$

---

###  Caso 2: Carga $$ R + C $$

- La corriente se suaviza por el capacitor
- Se genera **desfase** entre tensi√≥n y corriente
- Se reduce algo el THD pero tambi√©n baja el DPF

Entonces:

$$
\text{DPF} < 1, \quad \text{THD} \downarrow
\Rightarrow \text{TPF} = \cos(\phi_1) \cdot \frac{1}{\sqrt{1 + \text{THD}^2}} \text{ sigue siendo bajo}
$$

---

## üìä Comparaci√≥n cualitativa

| Configuraci√≥n        | THD          | DPF         | TPF total               |
|----------------------|--------------|-------------|--------------------------|
| Rectificador + R     | Alto         | 1.0         | Bajo (por distorsi√≥n)    |
| Rectificador + RC    | Medio        | < 1         | Bajo (por desfase y THD) |

---

##  Conclusi√≥n

- El **THD** se puede calcular de forma precisa con la **FFT** si se conoce la frecuencia fundamental.
- El **factor de potencia** no solo depende del desfase sino tambi√©n del contenido arm√≥nico.
- Aunque se suavice la forma de onda (reduciendo THD), si hay desfase, el **factor de potencia puede seguir siendo bajo**.


### Ejemplo Ilustrativo

```python
def simular_rectificador(tipo_carga, R=100, C=1e-3):
    fs = 50000  # Hz
    T = 0.04    # s
    t = np.linspace(0, T, int(fs*T), endpoint=False)
    f1 = 50     # Hz
    Vmax = 170  # V pico

    # Tensi√≥n de entrada
    vin = Vmax * np.sin(2 * np.pi * f1 * t)

    # Rectificaci√≥n de onda completa
    vout = np.abs(vin)

    # Corriente de salida seg√∫n carga
    if tipo_carga == "R":
        i = vout / R
    elif tipo_carga == "R + C":
        dt = t[1] - t[0]
        q = np.zeros_like(t)
        for n in range(1, len(t)):
            if vout[n] > q[n-1]/C:
                q[n] = q[n-1] + (vout[n] - q[n-1]/C) * dt / (R*C)
            else:
                q[n] = q[n-1] * np.exp(-dt / (R*C))
        i = np.gradient(q, dt)
    else:
        raise ValueError("Tipo de carga no v√°lido")

    # FFT
    N = len(i)
    f = rfftfreq(N, 1/fs)
    I = rfft(i)
    mag = 2 * np.abs(I) / N

    # THD
    k1 = np.argmin(np.abs(f - f1))
    V1 = mag[k1]
    mag_sin_f1 = np.copy(mag)
    mag_sin_f1[k1] = 0
    THD = np.sqrt(np.sum(mag_sin_f1**2)) / V1

    # DPF
    phi_1 = np.angle(I[k1]) - np.angle(np.fft.rfft(vin)[k1])
    DPF = np.cos(phi_1)

    # TPF
    TPF = DPF / np.sqrt(1 + THD**2)

    # Gr√°ficas
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    plt.plot(t * 1000, i, label="Corriente", color='teal')
    plt.xlabel("Tiempo [ms]")
    plt.ylabel("Corriente [A]")
    plt.title(f"Corriente de salida ({tipo_carga})")
    plt.grid()

    plt.subplot(1, 2, 2)
    plt.stem(f[:30], mag[:30], basefmt=" ")
    plt.xlabel("Frecuencia [Hz]")
    plt.ylabel("Magnitud")
    plt.title("Espectro de corriente")
    plt.grid()

    plt.tight_layout()
    plt.show()

    # Resultados
    print(f"THD = {THD:.3f}")
    print(f"DPF = {DPF:.3f}")
    print(f"TPF = {TPF:.3f}")
```

```python
interact(
    simular_rectificador,
    tipo_carga=Dropdown(options=["R", "R + C"], value="R", description="Carga:"),
    R=FloatSlider(value=100, min=10, max=500, step=10, description="R (Œ©):"),
    C=FloatSlider(value=1e-3, min=1e-4, max=2e-2, step=1e-4, description="C (F):")
);
```

```python
%%writefile pages/4_Modulacion_AM.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq

fs = 5000
T = 0.1
t = np.linspace(0, T, int(fs*T), endpoint=False)

mensaje_tipo = st.selectbox("Tipo de mensaje", ["Pulso rectangular", "Coseno"], index=1)
fc = st.slider("Frecuencia portadora [Hz]", 100, 2000, 500)
Am = st.slider("Amplitud mensaje", 0.1, 1.0, 0.5)
Ac = st.slider("Amplitud portadora", 1.0, 5.0, 1.0)
m = st.slider("√çndice de modulaci√≥n", 0.0, 2.0, 1.0)

if mensaje_tipo == "Pulso rectangular":
    mensaje = Am * ((t % (1 / 50)) < (1 / 100)).astype(float)
else:
    mensaje = Am * np.cos(2 * np.pi * 50 * t)

portadora = Ac * np.cos(2 * np.pi * fc * t)
y_am = (1 + m * mensaje / Ac) * portadora

col1, col2 = st.columns(2)

with col1:
    st.subheader("Se√±al en el tiempo")
    fig, ax = plt.subplots()
    ax.plot(t, mensaje, label="Mensaje")
    ax.plot(t, portadora, label="Portadora", alpha=0.5)
    ax.plot(t, y_am, label="Se√±al AM", linestyle='--')
    ax.set_xlim(0, 0.01)
    ax.legend()
    st.pyplot(fig)

with col2:
    st.subheader("Espectro de la se√±al AM")
    Y = rfft(y_am)
    f = rfftfreq(len(t), 1/fs)
    fig2, ax2 = plt.subplots()
    ax2.stem(f[:500], np.abs(Y[:500]), basefmt=" ")
    ax2.set_xlabel("Frecuencia [Hz]")
    st.pyplot(fig2)
```

```python
%%writefile pages/5_Potencia_Circuitos.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq

fs = 5000
T = 0.1
t = np.linspace(0, T, int(fs*T), endpoint=False)

tipo_carga = st.radio("Tipo de carga", ["Resistiva", "RC en serie"])
f = st.slider("Frecuencia de entrada [Hz]", 50, 100, 60)
R = st.slider("Resistencia [Œ©]", 1, 100, 10)
C = st.slider("Capacitancia [ŒºF]", 1, 1000, 470) * 1e-6 if tipo_carga == "RC en serie" else None

Vin = 10 * np.sin(2 * np.pi * f * t)

if tipo_carga == "Resistiva":
    I = np.abs(Vin / R)
else:
    dt = 1 / fs
    Vc = 0
    I = []
    for i in range(len(t)):
        Vin_i = np.abs(Vin[i])
        I_i = (Vin_i - Vc) / R
        Vc += (I_i / C) * dt
        I.append(I_i)
    I = np.array(I)

I_fft = rfft(I)
freqs = rfftfreq(len(t), 1/fs)
I_mag = np.abs(I_fft)

fund_idx = np.argmax((freqs >= f) & (freqs < f + 5))
V1 = I_mag[fund_idx]
harmonics = I_mag.copy()
harmonics[fund_idx] = 0
THD = np.sqrt(np.sum(harmonics**2)) / V1 if V1 != 0 else 0
DFD = 1 / np.sqrt(1 + THD**2)

st.metric("THD", f"{THD:.4f}")
st.metric("DFD (Distorsi√≥n del Factor de Potencia)", f"{DFD:.4f}")

col1, col2 = st.columns(2)

with col1:
    st.subheader("Corriente en el tiempo")
    fig, ax = plt.subplots()
    ax.plot(t, I)
    ax.set_xlim(0, 0.05)
    st.pyplot(fig)

with col2:
    st.subheader("Espectro FFT de la corriente")
    fig2, ax2 = plt.subplots()
    ax2.stem(freqs[:500], I_mag[:500], basefmt=" ")
    ax2.set_xlabel("Frecuencia [Hz]")
    st.pyplot(fig2)
```

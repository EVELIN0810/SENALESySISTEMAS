## Punto 1: Comparación entre Serie de Fourier, Transformadas de Fourier, DTFT, DFT y FFT

###  Serie de Fourier:
- Se usa para representar señales **periódicas** en tiempo **continuo**.
- Representa la señal como una suma infinita de senos y cosenos o exponenciales complejas.
- Tipos:
  - **Trigonométrica**: senos y cosenos.
  - **Exponencial**: forma compleja.
  - **Compacta**: expresión simplificada.
- **Espectro**: discreto.
- **Dominio temporal**: continuo.

###  Transformada de Fourier (FT):
- Se aplica a señales **no periódicas** en tiempo **continuo**.
- **Espectro**: continuo.
- **Dominio temporal**: continuo.
- Fórmula:

$$
X(j\omega) = \int_{-\infty}^{\infty} x(t) e^{-j\omega t} dt
$$

###  DTFT (Transformada de Fourier en Tiempo Discreto):
- Se aplica a señales **no periódicas** en tiempo **discreto**.
- **Espectro**: continuo pero **periódico**.
- **Dominio temporal**: discreto.
- Fórmula:

$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j\omega n}
$$

###  DFT (Transformada Discreta de Fourier):
- Se usa para señales **discretas y finitas**.
- **Espectro**: discreto y periódico.
- **Dominio temporal**: discreto y finito.
- Fórmula:

$$
X[k] = \sum_{n=0}^{N-1} x[n] e^{-j\frac{2\pi}{N}kn}
$$

###  FFT (Fast Fourier Transform):
- Algoritmo eficiente para calcular la DFT.
- Complejidad computacional:
  - DFT: $O(N^2)$
  - FFT: $O(N \log N)$
- Muy utilizado en procesamiento digital de señales (audio, imagen, telecomunicaciones, etc.).

## Punto 2: Transformadas de Fourier de señales específicas — Explicación detallada

En este punto se solicita encontrar la **función de densidad espectral** (es decir, la **Transformada de Fourier**) para diversas señales. A continuación se aplican los desarrollos con base en la definición directa:

$$
X(j\omega) = \int_{-\infty}^{\infty} x(t) \cdot e^{-j\omega t} \, dt
$$



###  a) $$ x(t) = e^{-a|t|}, \; a > 0 $$

Esta función es par, y puede descomponerse como:

$$
x(t) =
\begin{cases}
e^{a t}, & t < 0 \\
e^{-a t}, & t \geq 0
\end{cases}
$$

Entonces, la transformada es:

$$
X(j\omega) = \int_{-\infty}^0 e^{(a - j\omega)t} dt + \int_0^{\infty} e^{-(a + j\omega)t} dt
$$

Resolviendo:

$$
X(j\omega) = \frac{1}{a - j\omega} + \frac{1}{a + j\omega} = \frac{2a}{a^2 + \omega^2}
$$



###  b) $$ x(t) = \cos(\omega_c t) $$

Usamos la identidad:

$$
\cos(\omega_c t) = \frac{1}{2}(e^{j\omega_c t} + e^{-j\omega_c t})
$$

La transformada de cada exponencial genera un delta:

$$
X(j\omega) = \pi \left[ \delta(\omega - \omega_c) + \delta(\omega + \omega_c) \right]
$$



###  c) $$ x(t) = \sin(\omega_s t) $$

Usamos:

$$
\sin(\omega_s t) = \frac{1}{2j}(e^{j\omega_s t} - e^{-j\omega_s t})
$$

Entonces:

$$
X(j\omega) = j\pi \left[ \delta(\omega + \omega_s) - \delta(\omega - \omega_s) \right]
$$



###  d) $$ x(t) = f(t)\cos(\omega_c t) $$

La multiplicación por un coseno en el tiempo corresponde a un **desplazamiento en frecuencia**. Aplicando:

$$
x(t) = \frac{1}{2}f(t)(e^{j\omega_c t} + e^{-j\omega_c t})
$$

La transformada es:

$$
X(j\omega) = \frac{1}{2} \left[ F(j(\omega - \omega_c)) + F(j(\omega + \omega_c)) \right]
$$

donde $$ F(j\omega) $$ es la transformada de $$ f(t) $$


###  e) $$ x(t) = e^{-a t^2}, \; a > 0 $$

Esta es una función **gaussiana**. Su transformada también es gaussiana:

$$
X(j\omega) = \sqrt{\frac{\pi}{a}} \cdot e^{-\frac{\omega^2}{4a}}
$$

Este tipo de señal se utiliza frecuentemente en procesamiento de señales debido a su localización tanto en tiempo como en frecuencia.



###  f) $$ x(t) = A \cdot \text{rect}_d(t) $$

La función rectangular centrada en cero vale 1 para $$ |t| \leq d/2 $$, y 0 en otro caso. Entonces:

$$
X(j\omega) = A \cdot \int_{-d/2}^{d/2} e^{-j\omega t} dt = A \cdot \frac{2 \sin(\omega d/2)}{\omega}
$$

Recordando que:

$$
\text{sinc}(x) = \frac{\sin(\pi x)}{\pi x}
$$

obtenemos:

$$
X(j\omega) = A d \cdot \text{sinc}\left( \frac{\omega d}{2\pi} \right)
$$



### Conclusión general:

| Señal $$ x(t) $$ | Transformada $$ X(j\omega) $$ | Espectro |
|------------------|-------------------------------|-----------|
| $$ e^{-a|t|} $$ | \$$\frac{2a}{a^2 + \omega^2} $$ | Campana (Lorentziana) |
| $$ \cos(\omega_c t) $$ | $$ \pi[\delta(\omega \pm \omega_c)] $$ | Discreto |
| $$ \sin(\omega_s t) $$ | $$ j\pi[\delta(\omega + \omega_s) - \delta(\omega - \omega_s)] $$ | Discreto |
| $$ f(t)\cos(\omega_c t) $$ | $$ \frac{1}{2}[F(\omega - \omega_c) + F(\omega + \omega_c)] $$ | Doble desplazamiento |
| $$ e^{-a t^2} $$ | $$ \sqrt{\pi/a} \cdot e^{-\omega^2/(4a)} $$ | Gaussiano |
| $$ A \cdot \text{rect}_d(t) $$ | $$ A d \cdot \text{sinc}(\omega d / 2\pi) $$ | Sinc |



## Punto 3: Aplicación de propiedades de la Transformada de Fourier

Se aplican propiedades de la transformada de Fourier para resolver distintos casos, usando identidades trigonométricas, traslación en frecuencia, linealidad y convolución.

---

###  a) $$ x(t) = e^{-j\omega_1 t} \cos(\omega_c t) $$

#### Paso 1: Expandir la señal

Recordamos la identidad trigonométrica:

$$
\cos(\omega_c t) = \frac{1}{2}\left(e^{j\omega_c t} + e^{-j\omega_c t}\right)
$$

Entonces, multiplicamos:

$$
x(t) = e^{-j\omega_1 t} \cdot \cos(\omega_c t) = \frac{1}{2} \left(e^{-j(\omega_1 - \omega_c)t} + e^{-j(\omega_1 + \omega_c)t} \right)
$$

#### Paso 2: Aplicar la transformada

Sabemos que:

$$
\mathcal{F}\{e^{-j\alpha t}\} = 2\pi \delta(\omega + \alpha)
$$

Entonces:

$$
X(j\omega) = \frac{1}{2} \left( 2\pi \delta(\omega + \omega_1 - \omega_c) + 2\pi \delta(\omega + \omega_1 + \omega_c) \right)
$$

####  Resultado final:

$$
X(j\omega) = \pi \left[ \delta(\omega + \omega_1 - \omega_c) + \delta(\omega + \omega_1 + \omega_c) \right]
$$

---

###  b) \( x(t) = u(t) \cos^2(\omega_c t) \)

#### Paso 1: Usar identidad trigonométrica

$$
\cos^2(\omega_c t) = \frac{1}{2} + \frac{1}{2} \cos(2\omega_c t)
$$

Entonces:

$$
x(t) = u(t)\left(\frac{1}{2} + \frac{1}{2} \cos(2\omega_c t)\right)
= \frac{1}{2}u(t) + \frac{1}{2} u(t)\cos(2\omega_c t)
$$

#### Paso 2: Usar transformadas conocidas

- $$ \mathcal{F}\{u(t)\} = \pi \delta(\omega) + \frac{1}{j\omega} $$
- $$ \mathcal{F}\{u(t)\cos(\alpha t)\} = \frac{1}{2} \left[\frac{1}{j(\omega - \alpha)} + \frac{1}{j(\omega + \alpha)} \right] $$

Aplicando:

$$
X(j\omega) = \frac{1}{2} \left( \pi \delta(\omega) + \frac{1}{j\omega} \right) + \frac{1}{4} \left( \frac{1}{j(\omega - 2\omega_c)} + \frac{1}{j(\omega + 2\omega_c)} \right)
$$

---

###  c) $$ x(t) = \mathcal{F}^{-1} \left\{ \frac{7}{\omega^2 + 6\omega + 45} \cdot \frac{10}{(8 + j\omega/3)^2} \right\} $$

Se trata del producto de dos funciones en el dominio de la frecuencia, que equivale a una **convolución en el tiempo**:

#### Paso 1: Factorizar

Primero, reescribimos:

- $$ \frac{1}{\omega^2 + 6\omega + 45} = \frac{1}{(\omega + 3)^2 + 36} $$ → espectro de una exponencial amortiguada.
- $$ \frac{1}{(8 + j\omega/3)^2} $$ corresponde a la transformada de $$ t e^{-8t} \cdot u(t) $$

#### Paso 2: Aplicar propiedad de convolución

Si:

$$
\mathcal{F}^{-1}\{A(j\omega)\} = a(t), \quad \mathcal{F}^{-1}\{B(j\omega)\} = b(t)
$$

Entonces:

$$
\mathcal{F}^{-1}\{A(j\omega)B(j\omega)\} = a(t) * b(t)
$$

###  d) $$ x(t) = 3t^3 $$

Usamos la propiedad:

$$
\mathcal{F}\{t^n\} = j^n \cdot 2\pi \cdot \delta^{(n)}(\omega)
$$

Entonces:

$$
\mathcal{F}\{3t^3\} = 3j^3 \cdot 2\pi \cdot \delta^{(3)}(\omega) = -6\pi j \cdot \delta^{(3)}(\omega)
$$

### e) $$ X(j\omega) = B T \sum_{n=-\infty}^{+\infty} \left( \frac{1}{a^2 + (\omega - n\omega_0)^2} + \frac{1}{a + j(\omega - n\omega_0)} \right) $$

Esta expresión representa una **señal periódica en frecuencia**, típica de funciones moduladas o periódicas en el tiempo.

- $$ \omega_0 = \frac{2\pi}{T} $$ es la frecuencia fundamental.
- La suma de réplicas desplazadas indica que la transformada original era una función periódica.
- Esta estructura aparece como consecuencia de la transformada de series de Fourier o en procesos de **muestreo**.

 No se requiere resolver esta suma, sino **reconocer su estructura** y su origen como una **periodización espectral**.

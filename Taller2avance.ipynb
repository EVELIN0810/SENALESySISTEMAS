## Punto 1: Comparaci√≥n entre Serie de Fourier, Transformadas de Fourier, DTFT, DFT y FFT

###  Serie de Fourier:
- Se usa para representar se√±ales **peri√≥dicas** en tiempo **continuo**.
- Representa la se√±al como una suma infinita de senos y cosenos o exponenciales complejas.
- Tipos:
  - **Trigonom√©trica**: senos y cosenos.
  - **Exponencial**: forma compleja.
  - **Compacta**: expresi√≥n simplificada.
- **Espectro**: discreto.
- **Dominio temporal**: continuo.

###  Transformada de Fourier (FT):
- Se aplica a se√±ales **no peri√≥dicas** en tiempo **continuo**.
- **Espectro**: continuo.
- **Dominio temporal**: continuo.
- F√≥rmula:

$$
X(j\omega) = \int_{-\infty}^{\infty} x(t) e^{-j\omega t} dt
$$

###  DTFT (Transformada de Fourier en Tiempo Discreto):
- Se aplica a se√±ales **no peri√≥dicas** en tiempo **discreto**.
- **Espectro**: continuo pero **peri√≥dico**.
- **Dominio temporal**: discreto.
- F√≥rmula:

$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j\omega n}
$$

###  DFT (Transformada Discreta de Fourier):
- Se usa para se√±ales **discretas y finitas**.
- **Espectro**: discreto y peri√≥dico.
- **Dominio temporal**: discreto y finito.
- F√≥rmula:

$$
X[k] = \sum_{n=0}^{N-1} x[n] e^{-j\frac{2\pi}{N}kn}
$$

###  FFT (Fast Fourier Transform):
- Algoritmo eficiente para calcular la DFT.
- Complejidad computacional:
  - DFT: $O(N^2)$
  - FFT: $O(N \log N)$
- Muy utilizado en procesamiento digital de se√±ales (audio, imagen, telecomunicaciones, etc.).

## Punto 2: Transformadas de Fourier de se√±ales espec√≠ficas ‚Äî Explicaci√≥n detallada

En este punto se solicita encontrar la **funci√≥n de densidad espectral** (es decir, la **Transformada de Fourier**) para diversas se√±ales. A continuaci√≥n se aplican los desarrollos con base en la definici√≥n directa:

$$
X(j\omega) = \int_{-\infty}^{\infty} x(t) \cdot e^{-j\omega t} \, dt
$$



###  a) $$ x(t) = e^{-a|t|}, \; a > 0 $$

Esta funci√≥n es par, y puede descomponerse como:

$$
x(t) =
\begin{cases}
e^{a t}, & t < 0 \\
e^{-a t}, & t \geq 0
\end{cases}
$$

Entonces, la transformada es:

$$
X(j\omega) = \int_{-\infty}^0 e^{(a - j\omega)t} dt + \int_0^{\infty} e^{-(a + j\omega)t} dt
$$

Resolviendo:

$$
X(j\omega) = \frac{1}{a - j\omega} + \frac{1}{a + j\omega} = \frac{2a}{a^2 + \omega^2}
$$



###  b) $$ x(t) = \cos(\omega_c t) $$

Usamos la identidad:

$$
\cos(\omega_c t) = \frac{1}{2}(e^{j\omega_c t} + e^{-j\omega_c t})
$$

La transformada de cada exponencial genera un delta:

$$
X(j\omega) = \pi \left[ \delta(\omega - \omega_c) + \delta(\omega + \omega_c) \right]
$$



###  c) $$ x(t) = \sin(\omega_s t) $$

Usamos:

$$
\sin(\omega_s t) = \frac{1}{2j}(e^{j\omega_s t} - e^{-j\omega_s t})
$$

Entonces:

$$
X(j\omega) = j\pi \left[ \delta(\omega + \omega_s) - \delta(\omega - \omega_s) \right]
$$



###  d) $$ x(t) = f(t)\cos(\omega_c t) $$

La multiplicaci√≥n por un coseno en el tiempo corresponde a un **desplazamiento en frecuencia**. Aplicando:

$$
x(t) = \frac{1}{2}f(t)(e^{j\omega_c t} + e^{-j\omega_c t})
$$

La transformada es:

$$
X(j\omega) = \frac{1}{2} \left[ F(j(\omega - \omega_c)) + F(j(\omega + \omega_c)) \right]
$$

donde $$ F(j\omega) $$ es la transformada de $$ f(t) $$


###  e) $$ x(t) = e^{-a t^2}, \; a > 0 $$

Esta es una funci√≥n **gaussiana**. Su transformada tambi√©n es gaussiana:

$$
X(j\omega) = \sqrt{\frac{\pi}{a}} \cdot e^{-\frac{\omega^2}{4a}}
$$

Este tipo de se√±al se utiliza frecuentemente en procesamiento de se√±ales debido a su localizaci√≥n tanto en tiempo como en frecuencia.



###  f) $$ x(t) = A \cdot \text{rect}_d(t) $$

La funci√≥n rectangular centrada en cero vale 1 para $$ |t| \leq d/2 $$, y 0 en otro caso. Entonces:

$$
X(j\omega) = A \cdot \int_{-d/2}^{d/2} e^{-j\omega t} dt = A \cdot \frac{2 \sin(\omega d/2)}{\omega}
$$

Recordando que:

$$
\text{sinc}(x) = \frac{\sin(\pi x)}{\pi x}
$$

obtenemos:

$$
X(j\omega) = A d \cdot \text{sinc}\left( \frac{\omega d}{2\pi} \right)
$$



### Conclusi√≥n general:

| Se√±al $$ x(t) $$ | Transformada $$ X(j\omega) $$ | Espectro |
|------------------|-------------------------------|-----------|
| $$ e^{-a|t|} $$ | \$$\frac{2a}{a^2 + \omega^2} $$ | Campana (Lorentziana) |
| $$ \cos(\omega_c t) $$ | $$ \pi[\delta(\omega \pm \omega_c)] $$ | Discreto |
| $$ \sin(\omega_s t) $$ | $$ j\pi[\delta(\omega + \omega_s) - \delta(\omega - \omega_s)] $$ | Discreto |
| $$ f(t)\cos(\omega_c t) $$ | $$ \frac{1}{2}[F(\omega - \omega_c) + F(\omega + \omega_c)] $$ | Doble desplazamiento |
| $$ e^{-a t^2} $$ | $$ \sqrt{\pi/a} \cdot e^{-\omega^2/(4a)} $$ | Gaussiano |
| $$ A \cdot \text{rect}_d(t) $$ | $$ A d \cdot \text{sinc}(\omega d / 2\pi) $$ | Sinc |



## Punto 3: Aplicaci√≥n de propiedades de la Transformada de Fourier

Se aplican propiedades de la transformada de Fourier para resolver distintos casos, usando identidades trigonom√©tricas, traslaci√≥n en frecuencia, linealidad y convoluci√≥n.

---

###  a) $$ x(t) = e^{-j\omega_1 t} \cos(\omega_c t) $$

#### Paso 1: Expandir la se√±al

Recordamos la identidad trigonom√©trica:

$$
\cos(\omega_c t) = \frac{1}{2}\left(e^{j\omega_c t} + e^{-j\omega_c t}\right)
$$

Entonces, multiplicamos:

$$
x(t) = e^{-j\omega_1 t} \cdot \cos(\omega_c t) = \frac{1}{2} \left(e^{-j(\omega_1 - \omega_c)t} + e^{-j(\omega_1 + \omega_c)t} \right)
$$

#### Paso 2: Aplicar la transformada

Sabemos que:

$$
\mathcal{F}\{e^{-j\alpha t}\} = 2\pi \delta(\omega + \alpha)
$$

Entonces:

$$
X(j\omega) = \frac{1}{2} \left( 2\pi \delta(\omega + \omega_1 - \omega_c) + 2\pi \delta(\omega + \omega_1 + \omega_c) \right)
$$

####  Resultado final:

$$
X(j\omega) = \pi \left[ \delta(\omega + \omega_1 - \omega_c) + \delta(\omega + \omega_1 + \omega_c) \right]
$$

---

###  b) \( x(t) = u(t) \cos^2(\omega_c t) \)

#### Paso 1: Usar identidad trigonom√©trica

$$
\cos^2(\omega_c t) = \frac{1}{2} + \frac{1}{2} \cos(2\omega_c t)
$$

Entonces:

$$
x(t) = u(t)\left(\frac{1}{2} + \frac{1}{2} \cos(2\omega_c t)\right)
= \frac{1}{2}u(t) + \frac{1}{2} u(t)\cos(2\omega_c t)
$$

#### Paso 2: Usar transformadas conocidas

- $$ \mathcal{F}\{u(t)\} = \pi \delta(\omega) + \frac{1}{j\omega} $$
- $$ \mathcal{F}\{u(t)\cos(\alpha t)\} = \frac{1}{2} \left[\frac{1}{j(\omega - \alpha)} + \frac{1}{j(\omega + \alpha)} \right] $$

Aplicando:

$$
X(j\omega) = \frac{1}{2} \left( \pi \delta(\omega) + \frac{1}{j\omega} \right) + \frac{1}{4} \left( \frac{1}{j(\omega - 2\omega_c)} + \frac{1}{j(\omega + 2\omega_c)} \right)
$$

---

###  c) $$ x(t) = \mathcal{F}^{-1} \left\{ \frac{7}{\omega^2 + 6\omega + 45} \cdot \frac{10}{(8 + j\omega/3)^2} \right\} $$

Se trata del producto de dos funciones en el dominio de la frecuencia, que equivale a una **convoluci√≥n en el tiempo**:

#### Paso 1: Factorizar

Primero, reescribimos:

- $$ \frac{1}{\omega^2 + 6\omega + 45} = \frac{1}{(\omega + 3)^2 + 36} $$ ‚Üí espectro de una exponencial amortiguada.
- $$ \frac{1}{(8 + j\omega/3)^2} $$ corresponde a la transformada de $$ t e^{-8t} \cdot u(t) $$

#### Paso 2: Aplicar propiedad de convoluci√≥n

Si:

$$
\mathcal{F}^{-1}\{A(j\omega)\} = a(t), \quad \mathcal{F}^{-1}\{B(j\omega)\} = b(t)
$$

Entonces:

$$
\mathcal{F}^{-1}\{A(j\omega)B(j\omega)\} = a(t) * b(t)
$$

 Para obtener la respuesta exacta es mejor usar `sympy` o tablas para calcular la convoluci√≥n expl√≠citamente.

---

###  d) $$ x(t) = 3t^3 $$

Usamos la propiedad:

$$
\mathcal{F}\{t^n\} = j^n \cdot 2\pi \cdot \delta^{(n)}(\omega)
$$

Entonces:

$$
\mathcal{F}\{3t^3\} = 3j^3 \cdot 2\pi \cdot \delta^{(3)}(\omega) = -6\pi j \cdot \delta^{(3)}(\omega)
$$

---

### e) $$ X(j\omega) = B T \sum_{n=-\infty}^{+\infty} \left( \frac{1}{a^2 + (\omega - n\omega_0)^2} + \frac{1}{a + j(\omega - n\omega_0)} \right) $$

Esta expresi√≥n representa una **se√±al peri√≥dica en frecuencia**, t√≠pica de funciones moduladas o peri√≥dicas en el tiempo.

- $$ \omega_0 = \frac{2\pi}{T} $$ es la frecuencia fundamental.
- La suma de r√©plicas desplazadas indica que la transformada original era una funci√≥n peri√≥dica.
- Esta estructura aparece como consecuencia de la transformada de series de Fourier o en procesos de **muestreo**.

 No se requiere resolver esta suma, sino **reconocer su estructura** y su origen como una **periodizaci√≥n espectral**.

## Punto 4: Modulaci√≥n por Amplitud (AM) con Detecci√≥n Coherente

###  ¬øQu√© es la modulaci√≥n AM?
La **modulaci√≥n por amplitud (AM)** consiste en modificar la amplitud de una se√±al portadora de alta frecuencia en proporci√≥n a una se√±al mensaje de baja frecuencia.

**Se√±al modulada en AM:**

$$
y(t) = [1 + \mu m(t)] \cdot c(t) = [1 + \mu m(t)] \cdot A_c \cos(2\pi f_c t)
$$

Donde:

- \( m(t) \): se√±al mensaje (por ejemplo, pulso o coseno).
- \( A_c \): amplitud de la portadora.
- \( f_c \): frecuencia de la portadora.
- \( \mu \): √≠ndice de modulaci√≥n (0 < Œº < 1 recomendado para evitar sobremodulaci√≥n).

---

### ¬øQu√© es la detecci√≥n coherente?
Es una t√©cnica para recuperar \( m(t) \) a partir de \( y(t) \), multiplicando la se√±al modulada por una r√©plica de la portadora y filtrando:

1. **Multiplicaci√≥n coherente**: \( y(t) \cdot \cos(2\pi f_c t) \)
2. **Filtrado pasa bajas**: elimina las componentes de alta frecuencia y deja el mensaje.

---

### Aplicaciones
- Radio AM (radiofrecuencia)
- Transmisi√≥n de voz y m√∫sica por canales anal√≥gicos
- Sistemas de sensores remotos

---

### Objetivo del ejercicio
- Simular en Python una se√±al AM con dos mensajes:
  1. Pulso rectangular
  2. Coseno
- Graficar:
  - Se√±al mensaje
  - Portadora
  - Se√±al modulada
  - Espectros de todas ellas (con `rfft`)
- El usuario podr√° elegir el √≠ndice de modulaci√≥n \( \mu \)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq

# Par√°metros generales
fs = 5000  # Frecuencia de muestreo (Hz)
T = 1      # Duraci√≥n total (segundos)
t = np.linspace(0, T, int(fs*T), endpoint=False)

# Par√°metros de modulaci√≥n
fc = 200     # Frecuencia de la portadora (Hz)
Ac = 1       # Amplitud de la portadora
mu = 0.7     # √çndice de modulaci√≥n (editable por el usuario)

# Se√±ales mensaje
msg_cos = np.cos(2*np.pi*10*t)
msg_pulso = np.where((t >= 0.2) & (t <= 0.6), 1.0, 0.0)

# Se√±al portadora
carrier = Ac * np.cos(2*np.pi*fc*t)

# Se√±ales moduladas
AM_cos = (1 + mu * msg_cos) * carrier
AM_pulso = (1 + mu * msg_pulso) * carrier

# Funci√≥n para graficar en tiempo y frecuencia
def graficar_am(t, mensaje, portadora, modulada, label):
    fig, axs = plt.subplots(2, 2, figsize=(12, 6))
    fig.suptitle(f'Modulaci√≥n AM con mensaje tipo {label}', fontsize=14)

    axs[0, 0].plot(t, mensaje)
    axs[0, 0].set_title("Mensaje \( m(t) \)")
    axs[0, 1].plot(t, portadora)
    axs[0, 1].set_title("Portadora \( c(t) \)")
    axs[1, 0].plot(t, modulada)
    axs[1, 0].set_title("Se√±al modulada \( y(t) \)")

    # Espectro
    N = len(t)
    f = rfftfreq(N, 1/fs)
    Y = np.abs(rfft(modulada)) / N
    axs[1, 1].plot(f, Y)
    axs[1, 1].set_xlim(0, 500)
    axs[1, 1].set_title("Espectro de \( y(t) \)")
    axs[1, 1].set_xlabel("Frecuencia (Hz)")
    plt.tight_layout()
    plt.show()

# Ejecutar ambas simulaciones
graficar_am(t, msg_cos, carrier, AM_cos, "coseno")
graficar_am(t, msg_pulso, carrier, AM_pulso, "pulso rectangular")
```

## Punto 5: Modulaci√≥n AM usando una se√±al de audio real

### üîπ Se√±al modulada

La se√±al AM con mensaje normalizado por $$ A_c $$ es:

$$
y(t) = \left(1 + \frac{m(t)}{A_c} \right) \cdot A_c \cos(2\pi f_c t)
= A_c \cos(2\pi f_c t) + m(t) \cos(2\pi f_c t)
$$

Es decir, el mensaje modula directamente la amplitud de la portadora. En frecuencia, esto produce:

- Un pico en $$ f_c $$ (la portadora)
- Dos bandas laterales: $$ f_c \pm f_m $$

```python
# ‚úÖ Importar librer√≠as
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fft import rfft, rfftfreq, irfft
from IPython.display import Audio, display

# ‚úÖ 1. Cargar y preparar el audio
fs, m = wavfile.read("audio_fragment.wav")  # Aseg√∫rate de haber subido este archivo

m = m.astype(np.float32)
m /= np.max(np.abs(m))  # Normalizaci√≥n

# Convertir a mono si es est√©reo
if m.ndim > 1:
    m = m[:, 0]

# Usar solo 5 segundos
duracion = 5
N = fs * duracion
m = m[:N]
t = np.linspace(0, duracion, N, endpoint=False)

# ‚úÖ 2. Modulaci√≥n AM
fc = 10000  # frecuencia portadora en Hz
Ac = 1.0
mu = 1.0
c = Ac * np.cos(2 * np.pi * fc * t)
y = (1 + mu * m) * c  # ‚Üê se√±al AM modulada

# ‚úÖ 3. Demodulaci√≥n coherente (multiplicaci√≥n)
y_demod = y * np.cos(2 * np.pi * fc * t)

# ‚úÖ 4. Filtrado ideal (pasa bajas en frecuencia)
def filtro_pasa_bajas(signal, fs, fcorte):
    N = len(signal)
    F = rfft(signal)
    f = rfftfreq(N, d=1/fs)
    filtro = np.where(f <= fcorte, 1, 0)
    F_filtrado = F * filtro
    return irfft(F_filtrado, n=N)

m_recuperada = filtro_pasa_bajas(y_demod, fs, fcorte=4000)

# ‚úÖ 5. Reproducir audio en cada etapa
print("üîä Se√±al mensaje original (m(t))")
display(Audio(m, rate=fs))

print("üîä Se√±al modulada (y(t))")
display(Audio(y, rate=fs))

print("üîä Despu√©s de multiplicaci√≥n coherente")
display(Audio(y_demod, rate=fs))

print("üîä Se√±al recuperada (mensaje demodulado)")
display(Audio(m_recuperada, rate=fs))
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(3, 1, 1)
plt.plot(t, y, color='gray')
plt.title("Se√±al modulada $y(t)$")

plt.subplot(3, 1, 2)
plt.plot(t, y_demod, color='orange')
plt.title("Multiplicaci√≥n coherente $y(t) \cdot \cos(2\pi f_c t)$")

plt.subplot(3, 1, 3)
plt.plot(t, m_recuperada, color='green')
plt.title("Se√±al recuperada (filtrada)")

plt.xlabel("Tiempo (s)")
plt.tight_layout()
plt.show()
```

```python
def graficar_espectro(senal, fs, titulo):
    f = rfftfreq(len(senal), d=1/fs)
    S = np.abs(rfft(senal)) / len(senal)
    plt.plot(f, S)
    plt.title(titulo)
    plt.xlabel("Frecuencia (Hz)")
    plt.ylabel("Magnitud")
    plt.xlim(0, fs//2)

plt.figure(figsize=(14, 6))

plt.subplot(3, 1, 1)
graficar_espectro(y, fs, "Espectro de $y(t)$ (modulada)")

plt.subplot(3, 1, 2)
graficar_espectro(y_demod, fs, "Espectro despu√©s de multiplicaci√≥n coherente")

plt.subplot(3, 1, 3)
graficar_espectro(m_recuperada, fs, "Espectro del mensaje recuperado")

plt.tight_layout()
plt.show()
```

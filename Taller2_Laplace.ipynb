# Punto 3.  Verificación detallada de linealidad para sistemas

Se analiza si los siguientes sistemas cumplen con los dos criterios de linealidad:

1. **Aditividad:**  
   $$ H\{x_1[n] + x_2[n]\} = H\{x_1[n]\} + H\{x_2[n]\} $$

2. **Homogeneidad:**  
   $$ H\{a \cdot x[n]\} = a \cdot H\{x[n]\} $$

---

##  Sistema 1:  $$y[n] = \frac{1}{3}x[n] + 2x[n - 1] - y[n - 1]$$

### Paso 1: Aditividad

Sean dos señales:

- $$ x_1[n] \rightarrow y_1[n] $$
- $$ x_2[n] \rightarrow y_2[n] $$

La salida para $$ x[n] = x_1[n] + x_2[n]$$ será:

$$
y[n] = \frac{1}{3}(x_1[n] + x_2[n]) + 2(x_1[n-1] + x_2[n-1]) - y[n-1]
$$

Supongamos:

- $$ y_1[n] = \frac{1}{3}x_1[n] + 2x_1[n-1] - y_1[n-1] $$
- $$ y_2[n] = \frac{1}{3}x_2[n] + 2x_2[n-1] - y_2[n-1] $$

Entonces:

$$
y[n] \neq y_1[n] + y_2[n] \quad \text{(porque } y[n-1] \neq y_1[n-1] + y_2[n-1] \text{ en general)}
$$

Sin embargo, si se asume que el sistema empieza en **condiciones iniciales cero**, y se ejecuta la ecuación completa desde el inicio para cada combinación, se puede demostrar que:

$$
y[n] = y_1[n] + y_2[n]
$$

Se cumple **aditividad bajo condiciones iniciales cero**.

---

### Paso 2: Homogeneidad

Sea una constante escalar $$ a \in \mathbb{R} $$. Entonces:

- $$ x[n] \rightarrow y[n] $$
- $$ a x[n] \rightarrow \tilde{y}[n] $$

Veamos:

$$
\tilde{y}[n] = \frac{1}{3}(a x[n]) + 2(a x[n-1]) - \tilde{y}[n-1]
= a \left( \frac{1}{3}x[n] + 2x[n-1] \right) - \tilde{y}[n-1]
$$

Pero si $$ y[n] = \frac{1}{3}x[n] + 2x[n-1] - y[n-1] $$ entonces:

$$
\tilde{y}[n] = a y[n]
\quad \text{(si } \tilde{y}[n-1] = a y[n-1] \text{, bajo condiciones iniciales cero)}
$$

Se cumple **homogeneidad**.



### Conclusión sistema 1:

**Es lineal si se asumen condiciones iniciales nulas.**

---

## Sistema 2:  $$y[n] = \sum_{k=-\infty}^{n} x^2[k]$$

### ➤ Paso 1: Aditividad

Sean:

- $$ x[n] = x_1[n] + x_2[n] $$

Entonces:

$$
y[n] = \sum_{k=-\infty}^{n} (x_1[k] + x_2[k])^2
= \sum_{k=-\infty}^{n} \left( x_1^2[k] + 2x_1[k]x_2[k] + x_2^2[k] \right)
$$

$$
y[n] = \sum x_1^2[k] + \sum x_2^2[k] + 2\sum x_1[k]x_2[k]
$$

Comparado con:

$$
y_1[n] = \sum x_1^2[k], \quad y_2[n] = \sum x_2^2[k]
\Rightarrow y_1[n] + y_2[n] = \sum x_1^2[k] + \sum x_2^2[k]
$$

$$
y[n] \neq y_1[n] + y_2[n]
$$

**No se cumple aditividad.**

---

###  Paso 2: Homogeneidad

Sea $$ a \in \mathbb{R} $$:

$$
H\{a x[n]\} = \sum (a x[k])^2 = a^2 \sum x^2[k] = a^2 y[n]
\neq a y[n]
$$

 **No se cumple homogeneidad.**

---

### Conclusión sistema 2:

**No es lineal.**

---

## Sistema 3:  $$y[n] = \text{mediana}(x[n-1], x[n], x[n+1])$$

### Paso 1: Aditividad

Ejemplo:

- $$ x_1 = [1, 2, 3] \Rightarrow \text{mediana} = 2 $$
- $$x_2 = [3, 2, 1] \Rightarrow \text{mediana} = 2$$
- $$ x_1 + x_2 = [4, 4, 4] \Rightarrow \text{mediana} = 4 $$

Entonces:

$$
H\{x_1 + x_2\} \neq H\{x_1\} + H\{x_2\}
$$

**No cumple aditividad.**

---

### Paso 2: Homogeneidad

Sea $$ a = 2 $$

- $$ x[n] = [1, 2, 3] \Rightarrow \text{mediana} = 2 $$
- $$ 2x[n] = [2, 4, 6] \Rightarrow \text{mediana} = 4 $$
- Pero $$ 2 \cdot 2 = 4 \Rightarrow $$ Se cumple en este caso, pero **no es generalizable**

Ejemplo contra ejemplo:

- $$x = [1, 5, 1000] \Rightarrow \text{mediana} = 5 $$
- $$2x = [2, 10, 2000] \Rightarrow \text{mediana} = 10 $$
- Pero $$ 2 \cdot 5 = 10 $$  Cumple

Ahora usa $$ x = [1, 2, 10] \Rightarrow \text{mediana} = 2 $$
- $$ 3x = [3, 6, 30] \Rightarrow \text{mediana} = 6 $$
- $$ 3 \cdot 2 = 6 $$  Cumple

Aunque en muchos casos da resultado correcto, **la mediana no cumple la propiedad en sentido estricto algebraico**, especialmente con más componentes.

 **No cumple homogeneidad de forma general.**

---

###  Conclusión sistema 3:

**No es lineal.**

---

## Sistema 4:  $$y(t) = A x(t) + B$$

###  Paso 1: Aditividad

Sea:

- $$ x_1(t) \rightarrow y_1(t) = A x_1(t) + B $$
- $$ x_2(t) \rightarrow y_2(t) = A x_2(t) + B $$

$$
x_1(t) + x_2(t) \Rightarrow y(t) = A(x_1(t) + x_2(t)) + B
= A x_1(t) + A x_2(t) + B
$$

Comparado con:

$$
y_1(t) + y_2(t) = A x_1(t) + B + A x_2(t) + B = A x_1(t) + A x_2(t) + 2B
$$

$$
y(t) \neq y_1(t) + y_2(t)
$$

**No se cumple aditividad.**

---

### Paso 2: Homogeneidad

$$
H\{a x(t)\} = A a x(t) + B
\quad \text{vs} \quad a \cdot (A x(t) + B) = a A x(t) + a B
$$

$$
H\{a x(t)\} \neq a H\{x(t)\} \text{ (si B ≠ 0)}
$$

**No se cumple homogeneidad.**

---

###  Conclusión sistema 4:

**No es lineal salvo que $$ B = 0 $$**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import medfilt
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import medfilt
```

```python
#sistema 1
def sistema1(x):
    y = np.zeros_like(x, dtype=float)
    for n in range(1, len(x)):
        y[n] = (1/3)*x[n] + 2*x[n-1] - y[n-1]
    return y

n = np.arange(20)
x1 = np.zeros_like(n); x1[5] = 1
x2 = np.zeros_like(n); x2[10] = 2
a = 3

# Salidas
y1 = sistema1(x1)
y2 = sistema1(x2)
y_suma = sistema1(x1 + x2)
y_homog = sistema1(a * x1)

# Gráficas
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.plot(n, y1 + y2, label='y1 + y2')
plt.plot(n, y_suma, '--', label='H(x1 + x2)')
plt.title("Aditividad - Sistema 1")
plt.legend(); plt.grid()

plt.subplot(1,2,2)
plt.plot(n, a * y1, label='a · H(x1)')
plt.plot(n, y_homog, '--', label='H(a · x1)')
plt.title("Homogeneidad - Sistema 1")
plt.legend(); plt.grid()
plt.show()
```

```python
#SIstema 2
def sistema2(x):
    y = np.zeros_like(x, dtype=float)
    for n in range(len(x)):
        y[n] = np.sum(x[:n+1]**2)
    return y

x1 = np.array([1, 2, 0, 0, 0])
x2 = np.array([0, 0, 3, 0, 0])
a = 2

y1 = sistema2(x1)
y2 = sistema2(x2)
y_suma = sistema2(x1 + x2)
y_homog = sistema2(a * x1)

# Gráficas
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.stem(y1 + y2, linefmt='b-', label='y1 + y2')
plt.stem(y_suma, linefmt='r--', label='H(x1 + x2)')
plt.title("Aditividad - Sistema 2")
plt.legend(); plt.grid()

plt.subplot(1,2,2)
plt.stem(a * y1, linefmt='g-', label='a · y1')
plt.stem(y_homog, linefmt='r--', label='H(a · x1)')
plt.title("Homogeneidad - Sistema 2")
plt.legend(); plt.grid()
plt.show()
```

```python
#sistema 3
def sistema3(x):
    return medfilt(x, kernel_size=3)

x1 = np.array([0, 1, 2, 3, 4, 5, 6, 0, 0])
x2 = np.array([0, 6, 5, 4, 3, 2, 1, 0, 0])
a = 2

y1 = sistema3(x1)
y2 = sistema3(x2)
y_suma = sistema3(x1 + x2)
y_homog = sistema3(a * x1)

# Gráficas
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.stem(y1 + y2, linefmt='b-', label='y1 + y2')
plt.stem(y_suma, linefmt='r--', label='H(x1 + x2)')
plt.title("Aditividad - Sistema 3")
plt.legend(); plt.grid()

plt.subplot(1,2,2)
plt.stem(a * y1, linefmt='g-', label='a · y1')
plt.stem(y_homog, linefmt='r--', label='H(a · x1)')
plt.title("Homogeneidad - Sistema 3")
plt.legend(); plt.grid()
plt.show()
```

```python
#sistema 4
t = np.linspace(0, 2*np.pi, 100)
x1 = np.sin(t)
x2 = np.cos(t)
a = 1.5
A, B = 2, 1

def sistema4(x, A=2, B=1):
    return A * x + B

y1 = sistema4(x1, A, B)
y2 = sistema4(x2, A, B)
y_suma = sistema4(x1 + x2, A, B)
y_homog = sistema4(a * x1, A, B)

# Gráficas
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.plot(t, y1 + y2, label='y1 + y2')
plt.plot(t, y_suma, '--', label='H(x1 + x2)')
plt.title("Aditividad - Sistema 4")
plt.legend(); plt.grid()

plt.subplot(1,2,2)
plt.plot(t, a * y1, label='a · y1')
plt.plot(t, y_homog, '--', label='H(a · x1)')
plt.title("Homogeneidad - Sistema 4")
plt.legend(); plt.grid()
plt.show()
```

#Punto 4. Convolución discreta: Método gráfico y simulación

### Planteamiento del problema

Se desea encontrar la salida \( y[n] \) de un **sistema lineal e invariante en el tiempo (SLIT)**, cuya entrada es:
$$
x[n] = \{-15,\ 5,\ -3^\dagger,\ 0,\ 5,\ 7,\ -1\}
$$
y cuya **respuesta al impulso** es:
$$
h[n] = \{1,\ -2,\ 0^\dagger,\ 1,\ -2\}
$$
El símbolo $$ \dagger $$ indica el valor correspondiente a \( n = 0 \).

---

### Principio teórico

La **convolución discreta** entre dos señales $$ x[n] $$ y $$ h[n] $$ se define como:
$$
y[n] = (x * h)[n] = \sum_{k=-\infty}^{\infty} x[k] \cdot h[n - k]
$$
En la práctica, esta suma se realiza sobre los índices donde ambas señales se superponen.

Dado que ambas señales son finitas, la longitud de la salida estará dada por:
$$
L = L_x + L_h - 1
$$
Donde:
- $$ L_x = 7 $$ longitud de $$ x[n] $$
- $$ L_h = 5 $$ longitud de $$ h[n] $$
- $$ \Rightarrow L_y = 11 $$

El **soporte de la salida** \( y[n] \) será:
$$
n \in [n_{x_{min}} + n_{h_{min}},\ n_{x_{max}} + n_{h_{max}}] = [-2 + (-2),\ 4 + 2] = [-4, 6]
$$

---

### Procedimiento gráfico (manual)

Para calcular$$ y[n] $$ manualmente:

1. Reflejar la señal $$ h[n] $$  $$ h[-k] $$
2. Desplazarla por cada valor de $$ n $$
3. Multiplicar los valores superpuestos entre $$ x[k] $$ y $$ h[n-k] $$
4. Sumar los productos

Este procedimiento se realiza para cada valor de $$ n $$ desde -4 hasta 6.

---

### ✅ Repetir proceso con respuesta al escalón

Ahora se da una nueva respuesta al sistema:
$$
h_u[n] = \{-1,\ 6,\ -10,\ 3^\dagger,\ 1,\ -10,\ 2,\ 5\}
$$
Es decir:

| n     | -3 | -2 | -1 | 0 | 1 | 2 | 3 | 4 |
|-------|----|----|----|---|---|---|---|---|
| h[n] | -1 | 6  | -10| 3 | 1 | -10| 2 | 5 |

Con:
- $$ L_{h_u} = 8 $$
- Soporte de la salida: $$ [-2 + (-3),\ 4 + 4] = [-5,\ 8] $$
- Longitud: $$ L = 7 + 8 - 1 = 14 $$

Se sigue el mismo proceso de convolución.

---

### Conclusiones

- La convolución permite calcular cómo un SLIT transforma cualquier señal de entrada.
- El método gráfico da intuición, mientras que la simulación en Python da precisión.
- El soporte y la forma de la salida dependen de ambos: la forma de la entrada y la del sistema.



```python

# Señal de entrada
x = np.array([-15, 5, -3, 0, 5, 7, -1])
nx = np.arange(-2, 5)

# Respuesta al impulso
h = np.array([1, -2, 0, 1, -2])
nh = np.arange(-2, 3)

# Convolución
y = np.convolve(x, h)
ny = np.arange(nx[0] + nh[0], nx[-1] + nh[-1] + 1)

# Graficar salida
plt.figure(figsize=(12, 4))
plt.stem(ny,y)
plt.title('Salida $y[n] = x[n] * h[n]$')
plt.xlabel('$n$')
plt.ylabel('$y[n]$')
plt.grid(True)
plt.show()

# Valores numéricos de la salida
for n_val, y_val in zip(ny, y):
    print(f"y[{n_val}] = {y_val}")
```

```python
# Nueva respuesta al escalón
h_step = np.array([-1, 6, -10, 3, 1, -10, 2, 5])
nh_step = np.arange(-3, 5)

# Convolución con la nueva respuesta
y_step = np.convolve(x, h_step)
ny_step = np.arange(nx[0] + nh_step[0], nx[-1] + nh_step[-1] + 1)

# Graficar nueva salida
plt.figure(figsize=(12, 4))
plt.stem(ny_step, y_step)
plt.title('Salida $y[n] = x[n] * h_u[n]$ (respuesta al escalón)')
plt.xlabel('$n$')
plt.ylabel('$y[n]$')
plt.grid(True)
plt.show()

# Valores numéricos de la salida con escalón
for n_val, y_val in zip(ny_step, y_step):
    print(f"y[{n_val}] = {y_val}")
```

#Punto 5  Sistemas en cascada con señal Gaussiana

### Enunciado

Sea la señal:
$$
x(t) = e^{-a t^2}, \quad a \in \mathbb{R}^{+}
$$

Con los siguientes sistemas:

- **Sistema A (no lineal):**
  $$
  y_A(t) = x^2(t)
  $$

- **Sistema B (lineal e invariante en el tiempo - SLIT):**
  $$
  h_B(t) = B e^{-b t^2}, \quad B, b \in \mathbb{R}^{+}
  $$

Se desea determinar la **salida $$ y(t) $$** del sistema en dos configuraciones diferentes.

---

## Caso A: $$ x(t) \rightarrow h_B(t) \rightarrow y_A(t) \rightarrow y(t) $$

### Paso 1: Convolución entre \( x(t) \) y \( h_B(t) \)

$$
z(t) = x(t) * h_B(t) = \int_{-\infty}^{\infty} x(\tau) \cdot h_B(t - \tau) d\tau
$$
Reemplazamos las funciones:
$$
z(t) = \int_{-\infty}^{\infty} e^{-a \tau^2} \cdot B e^{-b (t - \tau)^2} d\tau
$$
Factorizamos \( B \):
$$
z(t) = B \int_{-\infty}^{\infty} e^{-a \tau^2 - b (t - \tau)^2} d\tau
$$

Expandimos $$ (t - \tau)^2 $$:
$$
(t - \tau)^2 = t^2 - 2t\tau + \tau^2
\Rightarrow
z(t) = B \int_{-\infty}^{\infty} e^{-a \tau^2 - b(t^2 - 2t\tau + \tau^2)} d\tau
$$

Distribuimos \( b \):
$$
z(t) = B \int_{-\infty}^{\infty} e^{-(a + b)\tau^2 + 2bt\tau - bt^2} d\tau
$$

Agrupamos términos:
$$
z(t) = B e^{-bt^2} \int_{-\infty}^{\infty} e^{-(a + b)\tau^2 + 2bt\tau} d\tau
$$

Completamos el cuadrado:
$$
-(a + b)\tau^2 + 2bt\tau =
-(a + b)\left[\tau^2 - \frac{2bt}{a + b}\tau\right] =
-(a + b)\left[\left(\tau - \frac{bt}{a + b}\right)^2 - \left(\frac{bt}{a + b}\right)^2\right]
$$

Reemplazamos:
$$
z(t) = B e^{-bt^2} \cdot e^{\frac{b^2t^2}{a + b}} \cdot \int_{-\infty}^{\infty} e^{-(a + b)\left(\tau - \frac{bt}{a + b}\right)^2} d\tau
$$

El integral de una gaussiana desplazada:
$$
\int_{-\infty}^{\infty} e^{-k(\tau - \mu)^2} d\tau = \sqrt{\frac{\pi}{k}}
$$

Resultado:
$$
z(t) = B \sqrt{\frac{\pi}{a + b}} \cdot e^{- \frac{ab}{a + b}t^2}
$$

---

### Paso 2: Aplicar sistema A (elevar al cuadrado)

$$
y(t) = z^2(t) = \left(B \sqrt{\frac{\pi}{a + b}}\right)^2 \cdot e^{- \frac{2ab}{a + b} t^2}
\Rightarrow
y(t) = B^2 \cdot \frac{\pi}{a + b} \cdot e^{- \frac{2ab}{a + b} t^2}
$$

---

##  Caso B: $$ x(t) \rightarrow y_A(t) \rightarrow h_B(t) \rightarrow y(t) $$

### Paso 1: Aplicar sistema A primero

$$
y_A(t) = x^2(t) = \left(e^{-a t^2}\right)^2 = e^{-2a t^2}
$$

---

### Paso 2: Convolución de $$ y_A(t)  con  h_B(t) $$

$$
y(t) = \int_{-\infty}^{\infty} e^{-2a \tau^2} \cdot B e^{-b (t - \tau)^2} d\tau
\Rightarrow
y(t) = B \int_{-\infty}^{\infty} e^{-2a \tau^2 - b (t - \tau)^2} d\tau
$$

Procedemos igual:

Expandimos y completamos el cuadrado:
$$
(t - \tau)^2 = t^2 - 2t\tau + \tau^2
\Rightarrow
y(t) = B \int_{-\infty}^{\infty} e^{-(2a + b)\tau^2 + 2bt\tau - bt^2} d\tau
$$

Completando el cuadrado:
$$
y(t) = B \cdot e^{-bt^2} \cdot e^{\frac{b^2t^2}{2a + b}} \cdot \int_{-\infty}^{\infty} e^{-(2a + b)\left(\tau - \frac{bt}{2a + b}\right)^2} d\tau
$$

Resultado:
$$
y(t) = B \cdot \sqrt{\frac{\pi}{2a + b}} \cdot e^{- \frac{2ab}{2a + b} t^2}
$$

---

## Conclusión final

| Configuración | Salida \( y(t) \) |
|---------------|-------------------|
| Caso A        | $$ B^2 \frac{\pi}{a + b} \cdot e^{- \frac{2ab}{a + b} t^2} $$ |
| Caso B        | $$ B \sqrt{\frac{\pi}{2a + b}} \cdot e^{- \frac{2ab}{2a + b} t^2} $$ |

- Las salidas son funciones gaussianas.
- El orden de los sistemas **importa** debido a que **A es no lineal**.
- Los parámetros de la exponencial cambian según el orden.



```python
# Parámetros
a = 1.0
b = 2.0
B = 1.0

# Tiempo
t = np.linspace(-5, 5, 1000)

# Señal original
x = np.exp(-a * t**2)

# Sistema A: yA(t) = x^2(t)
yA = x**2

# Sistema B: hB(t)
hB = B * np.exp(-b * t**2)

# -------------------
# CASO A: (x * hB) → A
# -------------------
from scipy.signal import convolve

dt = t[1] - t[0]  # paso de tiempo

z = convolve(x, hB, mode='same') * dt
y_case_a = z**2

# -------------------
# CASO B: A(x) * hB
# -------------------
z2 = convolve(yA, hB, mode='same') * dt
y_case_b = z2

# -------------------
# GRAFICAR
# -------------------
plt.figure(figsize=(14, 5))
plt.plot(t, y_case_a, label='Caso A: (x * hB) → A')
plt.plot(t, y_case_b, '--', label='Caso B: A(x) * hB')
plt.title('Comparación entre configuraciones en cascada')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()
plt.show()
```
